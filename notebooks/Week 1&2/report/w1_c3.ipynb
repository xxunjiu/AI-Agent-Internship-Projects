### 2. Difference between np.nan, Non and  .isnull()
| Feature | `None` | `np.nan` | `.isnull()` |
| :--- | :--- | :--- | :--- |
| **Origin** | Python Built-in | NumPy / IEEE 754 | Pandas Method |
| **Data Type** | `NoneType` | `float` | Returns Boolean mask |
| **Used For** | General objects, strings | Numeric/Math data | Finding both `None` & `nan` |
| **Equality** | `None == None` <br> `None is True` | `np.nan == np.nan` is `False` | N/A |
| **Pandas Casting** | Often converted to `nan` | Remains `nan` | N/A |

### 5.Bining continuous variables
##### &emsp;&emsp; Syntax: Equal width binning

```python
df['AgeBand'] = pd.cut(df['Age'], 5, labels=[1, 2, 3, 4, 5])
```

######  &emsp;&emsp; Function: Splites the mathematical range of the data into 5 equal intervals
#####  &emsp;&emsp; Syntax: Custom boundary binning

```python
# Boundaries: (0,5], (5,15], (15,30], (30,50], (50,80]
df['AgeBand'] = pd.cut(df['Age'], [0, 5, 15, 30, 50, 80], labels=[1, 2, 3, 4, 5])
```

###### &emsp;&emsp; Function: Cuts the age variable into specific uneven custom ranges
##### &emsp;&emsp; Syntax: equal frequency binning

```python
# Splits based on 10%, 30%, 50%, 70%, 90% quantiles
df['AgeBand'] = pd.qcut(df['Age'], [0, 0.1, 0.3, 0.5, 0.7, 0.9], labels=[1, 2, 3, 4, 5])
```

###### &emsp;&emsp;Function: distributes the number of samples as equally as possible across the 5 categories based on percentiles

### 7. Feature extraction from pure text
##### &emsp;&emsp; Syntax: ie:

```python
df['Title'] = df.Name.str.extract('([A-Za-z]+)\.', expand=False)
```

###### Explanation: &emsp;&emsp; egex Explanation: ([A-Za-z]+)\. matches a continuous sequence of one or more alphabetic characters ([A-Za-z]+) immediately followed by a literal period . (escaping the wildcard behavior of a dot). This isolates strings like "Mr" or "Mrs" from the full text pattern
