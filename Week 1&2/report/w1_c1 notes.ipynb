### Syntax explanation

#### A. pd.series(data)
#####  &emsp;&emsp;Behavior: row index labels(one dimension)
##### &emsp;&emsp;ie:

```python
sdata = {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}
example_1 = pd.Series(sdata)
```

#### B. pd.DataFrame(data)
##### &emsp;&emsp;Behavior: A 2D labeld structure which keys become colum names and default interger index is automaticlly generated for rows.
##### &emsp;&emsp;ie:

```python
data = {
    'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada', 'Nevada'],
    'year': [2000, 2001, 2002, 2001, 2002, 2003], 
    'pop': [1.5, 1.7, 3.6, 2.44, 2.9, 3.2]
}
example_2 = pd.DataFrame(data)
```
