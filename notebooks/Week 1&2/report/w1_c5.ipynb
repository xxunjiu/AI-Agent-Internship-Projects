## 1. Data Combining

### 1.1 `concat()`

#### Core Syntax

```python
pd.concat([df1, df2], axis=0)
pd.concat([df1, df2], axis=1)
```

#### Explanation

`concat()` is used to directly combine multiple DataFrames.

It can combine data vertically or horizontally.

#### Meaning of `axis`

| Syntax | Meaning | Result |
|---|---|---|
| `axis=0` | Combine by rows | Stack vertically; add more rows |
| `axis=1` | Combine by columns | Join horizontally; add more columns |


#### Horizontal Combining Formula

```python
new_df = pd.concat([df1, df2], axis=1)
```

Use this when two DataFrames have the same row order but different columns.

#### Example

```python
result_up = pd.concat([text_left_up, text_right_up], axis=1)
```

This means: combine `text_left_up` and `text_right_up` side by side.


#### Vertical Combining Formula

```python
new_df = pd.concat([df1, df2], axis=0, ignore_index=True)
```

Use this when two DataFrames have the same columns but different rows.

#### Example

```python
result = pd.concat([result_up, result_down], axis=0, ignore_index=True)
```

This means: stack `result_up` and `result_down` vertically into one complete DataFrame.

#### Important

```text
axis=0 → vertical combining → add rows
axis=1 → horizontal combining → add columns
```

`ignore_index=True` resets the index after combining, which prevents duplicated index values.

# 3. Data Combining

## 3.1 `pd.concat()`

`concat()` is used to directly combine multiple DataFrames.

It can combine DataFrames vertically or horizontally.

---

## Horizontal Combining

## Syntax

```python
new_df = pd.concat([df1, df2], axis=1)
```

## Explanation

`axis=1` means combining by columns.

This joins two DataFrames side by side.

## Use When

Use `axis=1` when:

- The two DataFrames have the same row order.
- The two DataFrames have different columns.
- You want to add columns horizontally.

## Example

```python
result_up = pd.concat([text_left_up, text_right_up], axis=1)
```

This means: combine `text_left_up` and `text_right_up` side by side.

## Key Idea

```text
axis=1 → horizontal combining → add columns
```

## Vertical Combining

## Syntax

```python
new_df = pd.concat([df1, df2], axis=0)
```

## Recommended Syntax

```python
new_df = pd.concat([df1, df2], axis=0, ignore_index=True)
```

## Explanation

`axis=0` means combining by rows.

This stacks one DataFrame under another DataFrame.

## Use When

Use `axis=0` when:

- The two DataFrames have the same column names.
- The two DataFrames contain different rows.
- You want to stack data vertically.

## Example

```python
result = pd.concat([result_up, result_down], axis=0, ignore_index=True)
```

This means: stack `result_up` and `result_down` vertically into one complete DataFrame.

## Why Use `ignore_index=True`?

```python
ignore_index=True
```

resets the index after combining.

This prevents duplicated index values.

## Key Idea

```text
axis=0 → vertical combining → add rows
axis=1 → horizontal combining → add columns
```


# 4. `join()`

## Syntax

```python
new_df = df1.join(df2)
```

## Explanation

`join()` is a DataFrame method.

It combines two DataFrames based on their index.

By default, it is used for horizontal combining.

## Use When

Use `join()` when:

- Two DataFrames have matching indexes.
- You want to add columns from one DataFrame to another.
- The row order is already aligned.

## Example

```python
result_up = text_left_up.join(text_right_up)
```

This means: attach the columns of `text_right_up` to the right side of `text_left_up` based on index.

## Similar Syntax

```python
text_left_up.join(text_right_up)
```

is similar to:

```python
pd.concat([text_left_up, text_right_up], axis=1)
```

## Key Idea

```text
join() → combine horizontally by index
```

# 5. `merge()`

`merge()` is used to combine DataFrames based on a common key column or index.

It is similar to SQL joins.

---

## 5.1 Merge by a Common Column

## Syntax

```python
new_df = pd.merge(df1, df2, on="key_column")
```

## Explanation

Use this when two DataFrames share a common column.

The common column is called the key.

## Example

```python
result = pd.merge(means, survived_sex, on="Sex")
```

This means: merge `means` and `survived_sex` using the common column `Sex`.

## 5.2 Merge by Index

## Syntax

```python
new_df = pd.merge(
    df1,
    df2,
    left_index=True,
    right_index=True
)
```

## Explanation

Use this when two DataFrames do not have a common column, but their indexes match.


## Example

```python
result_up = pd.merge(
    text_left_up,
    text_right_up,
    left_index=True,
    right_index=True
)
```

This means: merge `text_left_up` and `text_right_up` using their indexes.

## Parameter Explanation

```python
left_index=True
```

means use the index of the left DataFrame.

```python
right_index=True
```

means use the index of the right DataFrame.


# 6. `concat()` vs `join()` vs `merge()`

| Method | Syntax | Main Use |
|---|---|---|
| `concat()` | `pd.concat([df1, df2], axis=0/1)` | Directly combine DataFrames vertically or horizontally |
| `join()` | `df1.join(df2)` | Combine horizontally by index |
| `merge()` | `pd.merge(df1, df2, on="key")` | Combine by a common key column |
