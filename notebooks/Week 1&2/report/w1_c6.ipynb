
## 2.1 Handling Missing Values

### Explanation

Missing values need to be filled before building the model.

There are two common types of variables:

| Variable Type | Example | Common Filling Method |
|---|---|---|
| Categorical variable | `Sex`, `Cabin`, `Embarked` | Fill with `"NA"` or the most common value |
| Numerical variable | `Age`, `Fare` | Fill with mean, median, or mode |

### Syntax

```python
df['column_name'] = df['column_name'].fillna(value)
```

For numerical variables:

```python
df['column_name'] = df['column_name'].fillna(df['column_name'].mean())
```

For categorical variables:

```python
df['column_name'] = df['column_name'].fillna('NA')
```

## 3. Train-Test Split

### Explanation

The dataset is divided into two parts:

| Dataset | Purpose |
|---|---|
| Training set | Used to train the model |
| Testing set | Used to evaluate the model on unseen data |

### Syntax

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    stratify=y,
    random_state=0
)
```

### Important Parameters

| Parameter | Meaning |
|---|---|
| `X` | Input features |
| `y` | Target variable |
| `stratify=y` | Keeps the class ratio similar in training and testing sets |
| `random_state=0` | Makes the result reproducible |

## 4. Model Building

Model building means choosing a machine learning algorithm and training it with the training data.

This note uses two models:

- Logistic Regression
- Random Forest Classifier


## 4.1 Logistic Regression

### Explanation

Logistic regression is a classification model.

Although it has the word “regression,” it is used to predict categories, such as:

```text
Survived or not survived
Class 0 or class 1
Yes or no
```

It predicts the probability of each class and then chooses the class with the higher probability.


### Syntax

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

To check the model score:

```python
model.score(X_train, y_train)
model.score(X_test, y_test)
```

## 4.2 Random Forest Classifier

### Explanation

Random forest is a tree-based classification model.

It combines many decision trees to make a more stable prediction.

Compared with a single decision tree, random forest usually reduces overfitting.

### Syntax

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

To control model complexity:

```python
model = RandomForestClassifier(
    n_estimators=100,
    max_depth=5
)
```

### Important Parameters

| Parameter | Meaning |
|---|---|
| `n_estimators` | Number of trees in the forest |
| `max_depth` | Maximum depth of each tree |
| `random_state` | Makes results reproducible |

## 5. Overfitting

### Explanation

Overfitting happens when a model performs very well on the training set but not as well on the testing set.

For example:

```text
Training score: 0.99
Testing score: 0.80
```

This means the model learned the training data too specifically.


For random forest, one way is:

```python
RandomForestClassifier(max_depth=5)
```
