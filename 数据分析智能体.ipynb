## 数据分析润色

```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI

load_dotenv("/Users/ritasun/test/.env")

llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.3,
    max_tokens=800
)

print("Model initialized successfully.")
```

```python
question = "请用三句话解释什么是LLM(Large Language Model)。"

response = llm.invoke(question)

print(response.content)
```

```python
analysis_result = """
第二周的数据分析主要基于 Titanic 数据集，从第一周的基础数据预览进一步深入到数据清洗、特征工程、业务问题分析和简单建模。

首先，我检查了数据中的缺失值。主要存在缺失值的字段包括 Age、Cabin 和 Embarked。
由于 Age 是数值型变量，因此使用中位数填补缺失值。
Embarked 是类别型变量，因此使用出现次数最多的类别进行填补。
由于 Cabin 缺失值较多，因此创建了一个新的变量 HasCabin，用来表示乘客是否有船舱信息，同时将 Cabin 中的缺失值填充为 Unknown。

接着，我对 Pclass、Sex 和 Embarked 等类别型变量进行了类型转换。
同时，我创建了两个新的特征：FamilySize 和 IsAlone。
FamilySize 表示乘客在船上的家庭成员总数，IsAlone 表示乘客是否独自登船。

在业务问题分析部分，我主要探索了不同因素与乘客生存率之间的关系。
分析结果显示，女性乘客的生存率高于男性乘客。
一等舱乘客的生存率高于三等舱乘客。
此外，独自登船的乘客生存率低于与家人同行的乘客。

最后，我使用 scikit-learn 建立了一个简单的逻辑回归模型。
模型使用 Pclass、Sex、Age、Fare、Embarked、FamilySize、IsAlone 和 HasCabin 等变量来预测乘客是否生存。
"""
```

```python
prompt = f"""
你是一名专业的商业数据分析师。

任务：
将下面这段数据分析总结，改写成一份清晰和正式的中文业务分析报告。

要求：
1. 保留原始含义
2. 不要编造新的数据、数字或结论
3. 语言正式、清楚、简洁
4. 分成三个部分：
   - 数据准备
   - 主要发现
   - 建模总结

原始分析总结：
{analysis_result}
"""

response = llm.invoke(prompt)

print(response.content)
```

```python
prompt1= f"""
请把下面的数据分析总结改写成一份更专业的数据分析报告，并补充一些业务洞察。

原始分析总结：
{analysis_result}
"""

response = llm.invoke(prompt1)
print(response.content)
```

## Hallucination
#### 案例1中出现了幻觉。由于没有明确限制模型只能基于原始分析总结进行改写。因此，模型可能会加入原始总结中没有出现的新结论。

```python
llm_high_temp = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=1,
    max_tokens=800
)
```

```python
prompt2 = f"""
请将下面的数据分析总结改写成适合学术展示的业务报告：

{analysis_result}
"""

for i in range(3):
    response2 = llm_high_temp.invoke(prompt2)
    print(f"第 {i+1} 次输出：")
    print(response2.content)
```

## Unstable
#### 在这个案例中，我将 temperature 设置为 1，并对同一个数据分析总结连续生成三次报告。虽然输入内容和任务要求相同，但模型输出可能在标题、段落结构、用词和表达方式上出现差异。

```python
prompt3 = f"""
请阅读下面的数据分析总结，并给出一些改进建议。

数据分析总结：
{analysis_result}
"""

response3 = llm.invoke(prompt3)

print(response3.content)
```

#### 这个案例其实并不能算是失败案例。但是如果更精进准确的回答，或许应该增加对回答长度或者回答语言之类的控制。
