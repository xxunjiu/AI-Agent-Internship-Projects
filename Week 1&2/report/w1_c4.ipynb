## 5. Line Plot

### ie：Survival Count by Fare

```python
fare_sur = df.groupby(["Fare"])["Survived"] \
             .value_counts()

fare_sur.plot(grid=True)

plt.title("Fare vs Survival")
plt.xlabel("Fare")
plt.ylabel("Count")

plt.show()
```

## 6. Sorting Data Before Visualization

### Sorted Version

```python
fare_sur = df.groupby(["Fare"])["Survived"] \
             .value_counts() \
             .sort_values(ascending=False)
```


## 7. Count Plot (Seaborn)

### ie：Survival by Passenger Class

```python
sns.countplot(
    x="Pclass",
    hue="Survived",
    data=df
)

plt.title("Passenger Class vs Survival")

plt.show()
```

## 8. Explanation

### x="Pclass"

Passenger class shown on x-axis.

### hue="Survived"

Different colors represent:
- Survived
- Not survived

### data=df

Uses DataFrame as dataset source.



## 9. KDE Plot (Density Plot)

### ie： Age Distribution by Survival

```python
facet = sns.FacetGrid(
    df,
    hue="Survived",
    aspect=3
)

facet.map(
    sns.kdeplot,
    "Age",
    shade=True
)

facet.set(
    xlim=(0, df["Age"].max())
)

facet.add_legend()
```
